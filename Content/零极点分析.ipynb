### 零极点分析

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
from matplotlib.patches import Circle
```

zplane 函数绘制线性系统的极点和零点。例如，在 -1/2 处为零且在 $0.9e^{−j2π0.3}$ 和 $0.9e^{j2π0.3}$ 处有一对复极点的简单滤波器为

```python
z = -0.5+0j
p1 = 0.9*np.exp(-2*np.pi*0.3j)

p2 = 0.9*np.exp(2*np.pi*0.3j)
```

```python
fig,ax = plt.subplots()
circle = Circle(xy=(0.0,0.0),radius=1,alpha=0.5,facecolor='white',ec='cornflowerblue',linestyle='--',)
ax.add_patch(circle)
#ax.set_clip_path(circle)
ax.scatter(np.real(z),np.imag(z), s=30, c= 'b')
#ax.plot(np.real(z),np.imag(z))
ax.scatter(np.real(p1),np.imag(p1),s=30, c='b', marker= 'x')
ax.scatter(np.real(p2),np.imag(p2),s=30, c='b', marker= 'x')
#circle = Circle(xy=(0.0,0.0),radius=1,alpha=0.9,facecolor='white',ec='cornflowerblue',linestyle='--')
#ax.add_patch(circle)
#ax.plot(np.real(p2),np.imag(p2))
ax.axis("equal")
ax.set_xlim(-1.2,1.2)
ax.set_ylim(-1.2,1.2)
ax.grid()
fig.savefig('零极点分析.png',dpi=500)
```

fvtool返回输入参数确定滤波器的幅频响应曲线，使用python进行滤波器分析时，可以使用signal.freqs_zpk函数来确认由零极点和系统增益确认的滤波器的幅频响应曲线，也可以达到相同的效果。

```python
z = np.array([z])
p = np.array([p1,p2])
k = 1
w, h = signal.freqs_zpk(z, p, k, worN = 4096)

fig,ax = plt.subplots()
ax.plot(w, 20 * np.log10(abs(h)))
ax.grid()
fig.savefig('幅频响应曲线.png',dpi=500)
```
